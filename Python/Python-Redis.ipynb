```python
import redis
import pymysql
import json
from datetime import datetime
import time
import random
```

```python
config_redis = {'host': 'localhost',
                'port': 6379,
                'db': 0,}

config_mysql = {'host': 'localhost',
                'port': 3306,
                'user': 'root',
                'password': 'test',
                'db':'testdb'}
```

```python
L = 40      # 起始溫度下限
U = 50      # 起始溫度上限
p = 0.2     # 溫度發生變化的機率
x = 2       # 溫度變化的異動範圍區間
```

```python
pool_redis = redis.ConnectionPool(**config_redis)
r = redis.StrictRedis(connection_pool = pool_redis)

conn = pymysql.connect(**config_mysql)
cur = conn.cursor()
conn.autocommit(True)
```

```python
def counter_data():
    global count
    
    count += random.randint(0, 5)
    
    pdDict = {'id': 'm1_s1',
              'dt':datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
              'v':count}
    print(pdDict)
    r.lpush('sensortype1', json.dumps(pdDict))
        
    sql = """INSERT INTO sensortype1 VALUES (%s, %s, %s)"""
    cur.execute(sql ,(pdDict['id'], pdDict['dt'], pdDict['v']))
```

```python
def temperature():
    global tpt
    
    rnd = 0 if random.random() < (1 - p) else random.random() * x * 2 - x
    tpt = round(tpt + rnd, 2)

    pdDict = {'id': 'm1_s2',
              'dt':datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
              'tpt':tpt}
    print(pdDict)
    r.lpush('sensortype2', json.dumps(pdDict))
        
    sql = """INSERT INTO sensortype2 VALUES (%s, %s, %s)"""
    cur.execute(sql ,(pdDict['id'], pdDict['dt'], pdDict['tpt']))
```

```python
count = 0
tpt = random.randint(L, U)

def produce():
    while True:
        counter_data()
        temperature()
        
        time.sleep(1)
```

```python
produce()
```

```python
def consumer(i):
    while r.llen('sensortype1') > i:
        r.rpop('sensortype1')
    
    while r.llen('sensortype2') > i:
        r.rpop('sensortype2')
```

```python
data_range = 50
consumer(data_range)
```
